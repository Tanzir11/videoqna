```python
from moviepy.editor import VideoFileClip

def extract_audio(video_path, output_audio_path):
    try:
        # Load the video file
        video_clip = VideoFileClip(video_path)

        # Extract audio
        audio_clip = video_clip.audio

        # Write audio to file
        audio_clip.write_audiofile(output_audio_path)

        print("Audio extracted successfully!")

    except Exception as e:
        print("Error:", str(e))

# Example usage
video_path = "/content/video1.mp4"  # Replace with your video file path
output_audio_path = "/content/audio_file/output_audio.wav"  # Replace with desired output audio file path

extract_audio(video_path, output_audio_path)
```

# Downloading the low weight hugging face embedding model

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
embeddings = HuggingFaceInstructEmbeddings(
    model_name="all-MiniLM-L6-v2",
    model_kwargs={"device": "cpu"})
```

# To store the embeddings

```python
from langchain_community.vectorstores import FAISS
```

```python
import pickle
def store_embeddings(docs, embeddings, sotre_name, path):
    vectorStore = FAISS.from_documents(docs, embeddings)
    with open(f"{path}/faiss_{sotre_name}.pkl", "wb") as f:
        pickle.dump(vectorStore,f)
```

# To load the stored embeddings

```python
def load_embeddings(sotre_name, path):
    with open(f"{path}/faiss_{sotre_name}.pkl", "rb") as f:
        VectorStore = pickle.load(f)
    return VectorStore
```

# Splitting the documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=500)
data = text_splitter.split_documents(docs)
```

# Importing necessassory modules

```python
from langchain.chat_models import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain.prompts import ChatPromptTemplate
from langchain.schema.output_parser import StrOutputParser
```

# prompts that can be updated

```python
prompts = """You're given the transcripts of 3 videos your job is to answer the questions based on the provided context which is the transcripts.
Context: {context}
question: {question}
Answer question based only on question parameters which are given.
Make sure to answer what has been asked.
"""
```

# Function to perform QnA

```python
def Qna(questions):
    db_instructEmbedd = load_embeddings(sotre_name='instructEmbeddings', path="Embedding_store")
    retriever = db_instructEmbedd.as_retriever(search_kwargs={"k": 3})
    temp1 = prompts
    prompt = ChatPromptTemplate.from_template(temp1)
    model = ChatOpenAI(model="gpt-4", temperature=0)
    chain = (
        {"context": retriever, "question": RunnablePassthrough()}
        | prompt
        | model
        | StrOutputParser()
        )

    data = chain.invoke(questions)
    return data
```
